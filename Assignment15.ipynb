```python
# Q1. What is Flask Framework? What are the advantages of Flask Framework?

Flask is a lightweight, micro web framework for Python. It is designed to make web development easy by providing essential tools and features while allowing developers
flexibility and scalability. Flask follows the WSGI (Web Server Gateway Interface) standard and uses Jinja2 as its template engine.

Advantages of Flask:
Lightweight: Minimalistic framework with no unnecessary features or dependencies.
Flexible: Provides the freedom to build applications the way you want.
Easy to Learn: Simple and beginner-friendly, making it great for small projects and prototyping.
```

```python
# Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.
from flask import Flask
app = Flask(__name__)

@app.route('/')
def hello_world():
    return 'Hello World!!'

if __name__ == '__main__':
    app.run(debug=True)
```

```python
# Q3. What is App routing in Flask? Why do we use app routes?

App routing refers to the process of mapping URLs to specific functions in a Flask application. It uses the @app.route() decorator to define routes.
Uses:
URL Mapping: Allows linking specific URLs to Python functions for handling requests.
Clean URL Design: Enables creation of user-friendly and meaningful URLs.
```

```python
# Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/” route to show the following details:
# Company Name: ABC Corporation
# Location: India
# Contact Detail: 999-999-9999
# Attach the screenshot of the output in Jupyter Notebook.

from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return '''
    <h1>Company Name: ABC Corporation</h1>
    <h2>Location: India</h2>
    <h3>Contact Detail: 999-999-9999</h3>
    '''

@app.route('/welcome')
def welcome():
    return '<h1>Welcome to ABC Corporation</h1>'

if __name__ == '__main__':
    app.run(debug=True)
```
